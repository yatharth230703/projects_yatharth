```python
from tensorflow.keras.layers import TextVectorization 
print("done")
```

```python
X= df['comment_text']
y = df[df.columns[2:]].values
```

```python
MAX_FEATURES = 200000
```

```python
vectorizer = TextVectorization(max_tokens =MAX_FEATURES,
                              output_sequence_length=1800,
                              output_mode='int')
```

```python
vectorizer.adapt(X.values)
print("adaptation complete")
```

```python
vectorizer('Hello')
```

```python
vectorized_text = vectorizer(X.values)
print("complete")
```

```python
vectorized_text
```

```python
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset = dataset.cache()
dataset = dataset.shuffle(16000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8) 
#an imp feature of prefetch is that it prevents bottlenecks 
print("done")
```

```python
dataset.as_numpy_iterator().next()
```

```python
batch_X, batch_y = dataset.as_numpy_iterator().next()
```

```python
print(batch_X.shape)
print(batch_y.shape)
```

```python
train = dataset.take(int(len(dataset)*0.7))
val = dataset.skip(int(len(dataset)*0.7)).take(int(len(dataset)*0.2))
test = dataset.skip(int(len(dataset)*0.9)).take(int(len(dataset)*0.1))
#here we are using the skip function in order to ensure the dataset isnt taking redundant values 
print("done")
```

```python
train_generator = train.as_numpy_iterator()
```

```python
train_generator.next()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional , Dense , Embedding
print("imported")
```

```python
model = Sequential()
#create the embedding layer, . embedding will be represented as max features(number of words)+1
#32 values long is an embedding

model.add(Embedding(MAX_FEATURES+1,32))
#lstm layer will have 32 different units, with the activation function being tanh (hyperbolic tan)
#bidirectional allows us to pass informatoin backwards and forwards (both directions) across our LSTM layers 
#the importance of implementing a bidirectional layer can be understood with the help of an example
'''
lets say the sentence is "i dont hate you"/
if read from right to left, the model is going to scan the word hate first and then cite the sentence 
as toxic, which it is not 
hence it is important to scan the sentence in both the directions (bidirectonally ) in order to obtain a correct estimate of its nature 
Bidirectional is used frequently in NLP

'''
model.add(Bidirectional(LSTM(32,activation ='tanh')))
model.add(Dense(128, activation= 'relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
#the final layer output should map to the target dataset , y in this case 
#the sigmoid activation is used here since we are expecting the output to be somewhere between 1 and 0
model.add(Dense(6,activation='sigmoid'))
print("done")
```

```python
model.compile(loss= 'BinaryCrossentropy' , optimizer = 'Adam')
print("done")
```

```python
model.summary()
```

```python
history = model.fit(train, epochs=10 , validation_data= val)
```

```python
input_text = vectorizer('You freaking suck')
res = model.predict(np.array([input_text]))
#same thing as 
#res = model.predict(np.expand_dims(input_text,0))
#we cannot simply pass model.predict(input text) because the model is expecting the input ot be in the shape of (none)*(no. of values in out sequence )* (lenght of embedding)
print(res)
```

```python
from tensorflow.keras.metrics import Precision, Recall , CategoricalAccuracy
print("done")
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
print("done")
```

```python
for batch in test.as_numpy_iterator():
    #unpacking the batch
    X_true, y_true = batch
    #Make a prediction
    yhat = model.predict(X_true)
    ##explore why we flattened these
    #flatten converts the dataframe into just 1 big array 
    y_true = y_true.flatten()
    yhat = yhat.flatten()
    pre.update_state(y_true, yhat)
    re.update_state(y_true, yhat)
    acc.update_state(y_true, yhat)
print("done")
```

```python
print(f'Precision: {pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```
